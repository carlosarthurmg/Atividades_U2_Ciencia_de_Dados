```python
import pandas as pd
import numpy as np
import re
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler
```

```python
# Nomes das colunas para o dataset Adult
column_names = ['age', 'workclass', 'fnlwgt', 'education', 'education-num',
                'marital-status', 'occupation', 'relationship', 'race', 'sex',
                'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'income']

```

```python

df = pd.read_csv('adult.data', header=None, names=column_names, na_values=' ?')
```

```python
#ver as informações
df.info()
```

```python
# Padronizar valores categóricos
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].str.strip()
```

```python
#Padronizar valores categóricos específicos
country_mapping = {
    ' United-States': 'United States',
    ' United States': 'United States',
}
```

```python
#Converter valores "?" para NaN
df.replace(' ?', np.nan, inplace=True)
```

```python
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
df_numeric = df[numeric_cols]
```

```python
#Min-Max Scaling
minmax_scaler = MinMaxScaler()
df_minmax = pd.DataFrame(minmax_scaler.fit_transform(df_numeric), columns=numeric_cols)
```

```python
#Z-score Standardization
zscore_scaler = StandardScaler()
df_zscore = pd.DataFrame(zscore_scaler.fit_transform(df_numeric), columns=numeric_cols)
```

```python
#Robust Scaling
robust_scaler = RobustScaler()
df_robust = pd.DataFrame(robust_scaler.fit_transform(df_numeric), columns=numeric_cols)
```

```python
print("\nResumo estatístico - Original:")
print(df_numeric.describe())
```

```python
print("\nResumo estatístico - Min-Max Scaling:")
print(df_minmax.describe())
```

```python
print("\nResumo estatístico - Z-Score Standardization:")
print(df_zscore.describe())
```

```python
print("\nResumo estatístico - Robust Scaling:")
print(df_robust.describe())
```
